```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# X range for standard normal
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)  # Standard normal PDF

# Point at which to draw vertical line
x_val = 1.66

# Plot the normal curve
plt.plot(x, y, label='Standard Normal PDF', color='black')

# Shade area under curve up to x_val
x_fill = np.linspace(-4, x_val, 1000)
y_fill = norm.pdf(x_fill)
plt.fill_between(x_fill, y_fill, color='skyblue', alpha=0.5, label=f'AUC (X ≤ {x_val})')

# Draw vertical line at x value
plt.axvline(x=x_val, color='red', linestyle='--', label=f'X = {x_val}')

# Labels and legend
plt.title('Standard Normal Distribution and AUC up to X = 1.66')
plt.xlabel('X')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)

# Show plot
plt.show()
```
